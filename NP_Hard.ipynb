## **1. PROBLEMAS NP- HARD**

### **¿Qué quiere decir que un problema NP-Hard?**

Los problemas NP("tiempo polinomial no determinista") es un conjunto de problemas que se pueden resolver en un tiempo polinómico.

NP-hard  es un conjunto de problemas de decisión, contiene problemas H tales que el problema L en NP puede ser transformado polinomialmente en H.

Son problemas con una complejidad comparable a los NP-Completos, pero no necesariamente pertenecen al conjunto NP. 


Formalmente, un problema A es NP-Hard, si cualquier problema NP es reducible a A en tiempo polinomial.


### **¿Qué quiere decir que $P\neq NP$?**


$P\neq NP$ es una **conjetura**. Es decir algo que se cree que es así, pero no está probado.

La letra P se refiere a aquellos cálculos que se realizan a través de un tiempo polinomial 

Sin embargo, existen problemas matemáticos que no son aplicables bajo el tiempo polinómico: son difíciles de calcular, pero fáciles de verificar, esto recibe el nombre de NP.




En el caso de que $P=NP$

significa que existe un algoritmo polinomial para resolver NP-hard, es se reduce a que todos los problemas NP se pueden resolver en tiempo polinomial 



Si la conjetura es $P\neq NP$ tendríamos esta imagen aunque decir esto es impreciso, ya que hasta al día de hoy no está probado

### **¿Cómo se prueba  que un problema es NP-Hard usando reducciones?**

Para probar que un problema A es NP-hard se debe encontrar un problema B que sea NP-Hard, si B es reducible a A, entonces se comprueba que es NP-hard



Para demostrar que un problema $B$ es **intratable** o **NP-Hard** debemos aplicar **reducciones:**

* Seleccionamos un problema $A$ que sabemos que es **NP-Hard**.
* Si el problema $A$ es **reducible** a $B$, entonces el problema $B$  es **NP-Hard** también.

### **Subset sum**

Este problema es un caso especial del problema de la mochila, se basa en encontrar el subconjunto de un conjunto de de números que logra la suma indicada en un inicio.

La mayoría de algoritmos que resuelven este problema se caracterizan por que
que no hacen uso de la estructura especial del grupo subyacente, sino que solo usan el hecho que hay n elementos que tienen que sumar a un elemento de destino.

La complejidad de la resolución de este problema depende de $N$ : *el número de variables de decisión* y $P$: *la precisión del problema, el número de valores de posiciones binaria que toma definir el problema*.

Este problema es **NP** porque es un problema de decisión y su defición es simple.

###**Vertex Cover**

Dado un gráfico no dirigido, el problema de la cobertura de vértices es encontrar la cobertura de vértices de tamaño mínimo .

**Input**: Gráfo no dirigido $G(v,e)$ y un tamaño k.

**Output**: La cobertura de vértices de tamaño mínimo .


vertexcover.png


**Vertex cover** de un grafo no dirigido es un conjunto de vértices tal que por cada borde $(V,E)$ del gráfo, tanto $V$ como $E$ estan en el **vertex cover**. Dado un gráfico no dirigido, el problema de la cobertura de vértices es encontrar la cobertura de vértices de tamaño mínimo .

Vertex cover es un problema **NP**, dado que cualquier solución puede ser verificada e tiempo polinomial con $n^2$ examinaciones a todos los bordes, para ver si sus puntos finales se encuentrar incluidos en el vertex cover propuesto.

Existen diferentes formás de solucionar este problema, recordemos que este es un problema NP-Hard, por ende muchas de estas son muy ineficientes, por esto una de las soluciones más comunes para este problema es el algoritmo que entrega un aproximado del vertex cover.

## **4.Descripción de algoritmos y tiempos de ejecución**

### **Subset sum**

Este algoritmo cuenta con más de un método de resolución, esta vez nos enfocaremos en el método con programación dinámica para resolver el problema en tiempo pseudo-polinomial.

Los pasos que sigue este algoritmo son:

1. Crea una matriz de tipo boolean (Se debe inicializar en False), la cual tiene un tamaño de:
    **(Suma objetivo + 1)  *  (Largo arreglo de datos + 1)**.


2. Recorremos el arreglo de datos y asignamos **"True"** a todas las filas de la columna 0. Luego, recorremos nuevamente el arreglo en el rango desde el 2do dato hasta la posición equivalente al dato ingresado como suma objetivo, aquí se asigna a todas las columnas de la fila 0 como **False**.

3. Comenzamos con el proceso de programación dinamica, contiene el primer ciclo que va en el rango de 1 hasta el largo del arreglo anidado con el segundo ciclo que va en el rango de 1 hasta la posición equivalente al dato ingresado como suma. Una vez iterando estos ciclos, contienen condiciones, como:



> a. Si en el número de columna que estamos es menor a la posición del arreglo (fila actual - 1), se le asigna a la matriz en las posiciones actuales el booleano que está en la posición(fila - 1)(columna).





> b. Por el contrario, si el número de columna es mayor o igual a la condición nombrada anteriormente, a las posiciones actuales de fila y columna, les asignamos el booleano que se encuentra en la matriz en la posición  en la posición(fila - 1)(columna) o en  (fila -1)(columna - arreglo[fila-1]).


Estas condiciones nos dicen que, si el elemento en la posición actual contiene un número más grande que la suma actual, se copiaran los resultados anteriores. Y también nos dicen que, si la suma actual es mayor que el número "esimo", se revisa si algún estado anterior la obtuvo que la suma fuera igual a j o si algún elemento ya pasó por la posición (columna - arreglo[fila-1]).

4. Finalmente, retornamos el boolean que se encuentra en la matriz en la posición (largoArreglo)(sumaObjetivo), la cual responde a **si hay un subconjunto que logre la suma objetiva o no.**


El tiempo de ejecución de este algoritmo es de:  O(sumaObjetivo∗largoArreglo).
Esta solución no cuenta como de tiempo polinomial en teoría de complejidad porque P-N no es polinomial en el tamaño del problema.

###**Aproximación Vertex Cover**

Como ya mencionamos anteriormente, el problema de Vertex Cover es un problema NP-Hard, por lo que no tenemos soluciónes en tiempo polinomial, pese a esto, se han creados diferentes algoritmos que nos entregan una aproximación del vertex cover de un gráfo.

A continuación explicamos como se puede realizar una de estas aproximaciones.

***Pasos a seguir***

1. Inicializar el resultado como {}.
2. Considerar un set de todos los bordes del gráfo. Sea E el set.
3. Hacer mientras E no este vacío:
   a. Elegir un borde arbitrario $(u, v)$ del set E y agregar $u$ y $v$ al   resultado.
   b. Eliminar todos los bordes de $E$ que sean incidentess con $u$ o $v$.
4. Retornar resultado.
