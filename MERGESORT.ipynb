# **MERGE SORT**

Es un algoritmo de ordenamiento, que aplica el método **divide y vencerás**, el cual consiste en dividir el arreglo original en subarreglos hasta que sea lo suficientemente pequeño, es decir, que llegue a su caso base, donde el tamaño de el subarreglo será 1. Luego estos subarreglos se comenzaran a ordenar para dar como resultado el arreglo original ordenado.

## **Algoritmo MergeSort**
Este algoritmo requiere un arreglo de entrada que contendrá una x cantidad de elementos, donde **x** será el** tamaño de el problema** y su salida será el mismo arreglo pero con su contenido ordenado


Mediante un ejemplo se mostrará como funciona, se requiere ordenar el siguiente arreglo de manera ascendente, es decir, de menor a mayor:
**[6, 5, 3, 9, 7, 4]**, el cual sería la **entrada** en este problema
y su **salida** esperada sería : ** [3, 4, 5, 6, 7, 9]** .

En este ejemplo el **tamaño de el arreglo** será: **6**.



```python
def Merge(arrIzq, arrDer,cont):

  i = 0
  j = 0
  k = 0
  arrOrdenado = []

  while(i < len(arrIzq) and j < len(arrDer)):
    auxIzq= arrIzq[i]
    auxDer = arrDer[j]
    if(auxIzq <= auxDer):
      arrOrdenado.append(auxIzq)
      i+=1
      cont[0] = cont[0] + 1
    else:
      arrOrdenado.append(auxDer)
      j+=1
      cont[0] = cont[0] + 1  
  while(i < len(arrIzq)):    
    arrOrdenado.append(arrIzq[i])
    i+= 1
    cont[0] = cont[0] + 1

  while(j < len(arrDer)):    
    arrOrdenado.append(arrDer[j])
    j+= 1
    cont[0] = cont[0] + 1

     

  return arrOrdenado      

def MergeSort(arreglo, cont, verbose):

  mitad = len(arreglo) // 2
  if(len(arreglo) == 1):
    return arreglo
  if(verbose == True):
    arrIzq = arreglo[:mitad]
    arrDer = arreglo[mitad:]
    print("Subarreglos: ", arrIzq, arrDer, "\n")
  else:
      arrIzq = arreglo[:mitad]
      arrDer = arreglo[mitad:]
    

  if(cont[1 != 0]):
    cont[1] = cont[1] + 1
  arrIzq = MergeSort(arrIzq,cont, verbose)
  arrDer = MergeSort(arrDer, cont,verbose)
  return Merge(arrIzq, arrDer,cont)


arreglo = [6, 5, 3, 9, 7, 4]
print("-- ARREGLO INICIAL --")
print(arreglo)
#posicion 0 : cantidad comparaciones, posicion : cantidad de llamadas recursivas
cont = [0,0]
arrOrdenado = MergeSort(arreglo,cont, verbose = False)
print("-- ARREGLO ORDENADO --")
print(arrOrdenado)
```



### **PASO A PASO**


**Variables importantes**


> Arreglo Izquierdo: Guarda los elementos de la mitad izquierda de el arreglo principal.


> Arreglo Derecho:  Guarda los elementos de la mitad derecha de el arreglo principal.


> Arreglo Ordenado: Guarda la combinación de el arreglo izquierdo y Derecho ordenado



El algoritmo Merge Sort es recursivo, es decir, se llama a si mismo hasta encontrar un elemento base, teniendo esto en consideración, diremos que los pasos principales hasta encontrar el caso base serán 1 y 2:

1.- Ordenar arreglo izquierdo

2.- Ordenar arreglo derecho

Cuando encontramos nuestro caso base, mezclamos ambos arreglos, el cual sería el 3 paso principal.

3.- Mezclar arreglos anteriores. 

**Ejemplo:**
Nuestro arreglo es [5, 3, 4, 3, 2].

Los primeros subarreglos serán arreglo izquierdo = [5, 3] y arreglo derecho = [4,3,2] y seguimos los pasos recursivos hasta llegar a el caso base.


**1.-** Ordenar arreglo izquierdo, pero como contiene dos datos, no es un caso base, volvemos a llamar a nuestra funcion MergeSort y en esta llamada nuestro arreglo izquierdo es = [3] y el arreglo derecho = [5], como cada subarreglo es un dígito estamos en nuestro **caso base **, por lo mismo ambos subarreglos estan ordenados y los paso 1 y 2 están listos, en el paso 3 se mezclan y ordenan ambos subarreglos para volver al subarreglo izquierdo principal, de esta forma la salida será [3,5] y el paso 1 está listo.

**2.-** Ordenar arreglo derecho, como vimos en el paso anterior, nuestro arreglo derecho es = [4, 3, 2] y al no ser un caso base, volvemos a llamar a MergeSort y repetimos los pasos, donde arreglo izquiero ahora es = [4] y arreglo derecho = [3, 2]. 

El paso 1 esta completo, debido que subarreglo izquierdo cumple el caso base y esta ordenado. En el paso 2, nuestro subarreglo derecho, vuelve a llamar a MergeSort porque no es un caso base, por lo que el subarreglo izquierdo en esta llamada es = [3] y el derecho = [2], donde ambos subarreglos estan ordenados y  el paso 3 de esta llamada a MergeSort tendría como salida = [2, 3] y asi el paso  esta completo . De la llamada anterior, el paso 3 recibe como entrada el subarreglo izquierdo = [4] y el subarreglo derecho = [2, 3], donde ambos por si solos están ordenados, así se ordenan de tal forma que el arreglo de salida será = [2 , 3, 4]

**3.-** Finalmente, ambos subarreglos principales están ordenados por separado, y en este último paso, se mezclan y se ordenan ambas mitades del arreglo original, donde el arreglo Ordenado será = [2, 3, 3, 4, 5].

## **TIEMPO DE EJECUCIÓN**
### **TEOREMA DE TIEMPO DE EJECUCÍON:**
El algoritmo Merge Sort, para el** mejor y peor caso**, tiene el** mismo tiempo de ejecución**, debido que en todas las situaciones siempre realiza las mismas iteraciones, es decir, si el arreglo esta ordenado de mayor a menor o de menor a mayor, en ambas ocasiones se irán diviendo cada vez en subarreglos.
El tiempo de ejecución es: **O(n log n)**
